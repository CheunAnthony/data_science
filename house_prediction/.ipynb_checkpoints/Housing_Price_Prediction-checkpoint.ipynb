# Data inputation
```
# Categorical values

For the categorical values, we are going to replace null values with the mode
```

```
# Numerical values

For the numerical values, if we have too many NA values, we are going to replace them with the median
```




````python
categorical_columns = categorical_features.to_list()
````

````python
for col in df[categorical_columns]:
  mode_value = df[col].mode()[0]
  df[col] = df[col].fillna(mode_value)
````

````python
numerical_colums = numerical_features.to_list()

for col in df[numerical_colums]:
  median_value = df[col].median()
  df[col] = df[col].fillna(median_value)
````

#Check if we still have some null values in our dataset

````python
df.isna().sum().sum()
````

#Apply label encoding on our categorical variables

````python
# Applying label encoding on categorical columns
categorical_columns = categorical_features.to_list()
label_encoder = LabelEncoder()

for col in categorical_columns:
  df[col] = label_encoder.fit_transform(df[col])
````

````python
#Verifying that all our categoricals columns have been transformed
df[categorical_columns].head()
````

# Feature Engineering

```
# Log transformation

Applying a log transformation on the column sale price to get a better distribution
```

```
# Meta features

Create new meta features by adding or dividing existing columns to get more context
```

````python
# Applying a log transformation to the target variable
df['SalePrice'] = np.log1p(df['SalePrice'])
````

````python
numerical_colums
````

````python
pd.Series(df['GarageCars']).value_counts()
````

````python
# Create new features
df["TotalArea"] = df["GrLivArea"] + df["TotalBsmtSF"]
df["TotalPorch"] = df["OpenPorchSF"] + df["EnclosedPorch"] + df["3SsnPorch"] + df["ScreenPorch"]
df['AgeOfHouse'] = df['YrSold'] - df['YearBuilt']
df['AgeSinceRemodel'] = df['YrSold'] - df['YearRemodAdd']
df['LotSizePerSqFt'] = df['LotArea'] / df['LotFrontage']
df['TotalBasementArea'] = df['BsmtFinSF1'] + df['BsmtFinSF2'] + df['BsmtUnfSF']
df['BasementQuality'] = df['BsmtQual']  + df['BsmtCond']
df['TotalBathrooms'] = df['BsmtFullBath'] + df['BsmtHalfBath'] + df['FullBath'] + df['HalfBath']
df['BathPerRoom'] = df['TotalBathrooms'] / df['TotRmsAbvGrd']
df['BedroomToBathRatio'] = df['BedroomAbvGr'] / df['TotalBathrooms']
#df['GarageSizePerCar'] = df['GarageArea'] / df['GarageCars']
df['OverallQualityCondition'] = df['OverallQual'] * df['OverallCond']
````

#Remove highly correlated features and useless columns

````python
corr_matrix = df.corr().abs()
threshold = 0.9
mask = np.triu(np.ones(corr_matrix.shape), k=1)
upper_triangle = corr_matrix.where(mask == 1)
to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]
df.drop(columns=to_drop,inplace=True)
````

````python
df.drop('Id',axis=1,inplace=True)
````

````python
exclude_value = 'SalePrice'
all_cols = df.columns.to_list()
feature_columns = [item for item in all_cols if item != exclude_value]
````

````python
# Scaling our data
scaler = RobustScaler()
df[feature_columns] = scaler.fit_transform(df[feature_columns])
````

#Train-Test Split

````python
x = df[feature_columns]
y = df['SalePrice']

X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
````

#Model Selection

## We have chosen the following models:

* LinearRegression

* DecisionTreeRegressor

* RandomForestRegressor


* GradientBoostingRegressor

* XGBRegressor

````python
df.isna().sum()
````

````python

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_val)

mae_lr = mean_absolute_error(y_val, y_pred_lr)
mse_lr = mean_squared_error(y_val, y_pred_lr)
r2_lr = r2_score(y_val, y_pred_lr)
print(f"Linear Regression - MAE: {mae_lr}, MSE: {mse_lr}, R^2: {r2_lr}")
````

````python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_val)

mae_dt = mean_absolute_error(y_val, y_pred_dt)
mse_dt = mean_squared_error(y_val, y_pred_dt)
r2_dt = r2_score(y_val, y_pred_dt)

print(f"Decision Tree - MAE: {mae_dt}, MSE: {mse_dt}, R^2: {r2_dt}")
````

````python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_val)

mae_rf = mean_absolute_error(y_val, y_pred_rf)
mse_rf = mean_squared_error(y_val, y_pred_rf)
r2_rf = r2_score(y_val, y_pred_rf)

print(f"Random Forest - MAE: {mae_rf}, MSE: {mse_rf}, R^2: {r2_rf}")
````

````python
gb_model = GradientBoostingRegressor(random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_val)

mae_gb = mean_absolute_error(y_val, y_pred_gb)
mse_gb = mean_squared_error(y_val, y_pred_gb)
r2_gb = r2_score(y_val, y_pred_gb)

print(f"Gradient Boosting - MAE: {mae_gb}, MSE: {mse_gb}, R^2: {r2_gb}")
````

````python
xgb_model = xgb.XGBRegressor(random_state=42)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_val)

mae_xgb = mean_absolute_error(y_val, y_pred_xgb)
mse_xgb = mean_squared_error(y_val, y_pred_xgb)
r2_xgb = r2_score(y_val, y_pred_xgb)

print(f"XGBoost - MAE: {mae_xgb}, MSE: {mse_xgb}, R^2: {r2_xgb}")
````

````python
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

rf_grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42), param_grid=param_grid, cv=5, n_jobs=-1)
rf_grid_search.fit(X_train, y_train)

print(f"Best Parameters: {rf_grid_search.best_params_}")
````
